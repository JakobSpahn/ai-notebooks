```python
%pylab inline
import numpy as np
from tqdm import trange
```

```python
def fetch(url):
    import requests, gzip, os, hashlib, numpy
    #Join file in /tmp (name is hash of url)
    fp = os.path.join("/tmp", hashlib.md5(url.encode('utf-8')).hexdigest())
    print(fp)
    print(os.path.isfile(fp))
    #Check if file already exists
    if os.path.isfile(fp):
        with open(fp, "rb") as f:
            dat = f.read()
    #If file doesn't exist, write 
    else:
        with open(fp, "wb") as f:
            dat = requests.get(url).content
            f.write(dat)
    #Return decompressed data
    return numpy.frombuffer(gzip.decompress(dat), dtype=np.uint8).copy()
X_train_d = fetch("http://yann.lecun.com/exdb/mnist/train-images-idx3-ubyte.gz")
Y_train_d = fetch("http://yann.lecun.com/exdb/mnist/train-labels-idx1-ubyte.gz")
X_test_d = fetch("http://yann.lecun.com/exdb/mnist/t10k-images-idx3-ubyte.gz")
Y_test_d = fetch("http://yann.lecun.com/exdb/mnist/t10k-labels-idx1-ubyte.gz")
```

```python
print(X_train_d.shape)
test = X_train_d[16:].reshape((-1,28,28))
print(test.shape)
Y_train_d.shape
```

```python
#Reshape: Images are of shape 28x28

#First 15 elements of initial array contain meta info
X_train = X_train_d[16:].reshape((-1,28,28))

#First 7 elements contain meta info
Y_train = Y_train_d[8:]

X_test = X_test_d[0x10:].reshape((-1,28,28))
Y_test = Y_test_d[8:]
```

```python
import torch
import torch.nn as nn
```

```python
#Model
class NrlNet(nn.Module):
    def __init__(self):
        super(NrlNet, self).__init__()
        self.l1 = nn.Linear(784,128)
        self.act = nn.ReLU()
        self.l2 = nn.Linear(128, 10)
    def forward(self, x):
        x = self.l1(x)
        x = self.act(x)
        x = self.l2(x)
        return x
model = NrlNet()
```

```python
#Train
model = NrlNet()
##Batch Size
BS = 128
##Epochs
EP = 1000
loss_function = nn.CrossEntropyLoss()
optim = torch.optim.Adam(model.parameters())
losses, accuracies = [], []

for i in (t := trange(EP)):
    #32 random samples from the 6000 elements in train
    samp = np.random.randint(0, X_train.shape[0], size=(BS))
    X = torch.tensor(X_train[samp].reshape(-1,28*28)).float()
    Y = torch.tensor(Y_train[samp]).long()
    
    optim.zero_grad()
    out = model(X)
    
    #Accuracy
    cat = torch.argmax(out, dim=1)
    accuracy = (cat == Y).float().mean()
    
    loss = loss_function(out,Y)
    loss.backward()
    optim.step()
    
    #Information and Graphs
    loss, accuracy = loss.item(), accuracy.item()
    losses.append(loss) 
    accuracies.append(accuracy)
    t.set_description("loss %.2f accuracy %.2f" % (loss, accuracy))

plt.ylim(-0.1,1.1)    
plot(losses)
plot(accuracies)
```

```python
#Evaluate
Y_test_preds = torch.argmax(model(torch.tensor(X_test.reshape((-1,28*28))).float()), dim=1).numpy()
accuracy_test = (Y_test == Y_test_preds).mean()
print("test accuracy: %.4f" % accuracy_test)
```
